```python
import regex as re

def extract_song_snippet(text):
    pattern = '(^|\n\n)(.*?)\n\n'
    search_results = re.findall(pattern, text, overlapped=True, flags=re.DOTALL)
    songs = [song[1] for song in search_results]
    print("Found {} songs in text".format(len(songs)))
    return songs

def save_song_to_abc(song, filename="tmp"):
    save_name = "{}.abc".format(filename)
    with open(save_name, "w") as f:
        f.write(song)
    return filename

def abc2wav(abc_file):
    suf = abc_file.rstrip('.abc')
    cmd = "abc2midi {} -o {}".format(abc_file, suf + ".mid")
    os.system(cmd)
    cmd = "timidity {}.mid -Ow {}.wav".format(suf, suf)
    return os.system(cmd)

def play_wav(wav_file):
    return ipythondisplay.Audio(wav_file)

def play_song(song):
    basename = save_song_to_abc(song)
    ret = abc2wav(basename + '.abc')
    if ret == 0: #did not suceed
        return play_wav(basename+'.wav')
    return None
```

###Creating training examples and targets
We will divide the text into example sequences that we'll use during training. Each input sequence that we feed into our RNN will contain `seq_length` characters from the text. We will also define a target sequence for each input sequence, which will be used in training the RNN to predict the next character. For each input, the corresponding target will contain the same length of text, except shifted one character to the right.

We'll break the text into chunks of `seq_length+1`. Suppose `seq_length` is 4 and our text is "Panda". Then, our input sequence is "Pand" and the target sequence is "anda".

The batch method will then let us convert this stream of character indices to sequences of the desired size.

```python
def get_batch(vectorized_songs, seq_length, batch_size):
  # the length of the vectorized songs string
  n = vectorized_songs.shape[0] - 1
  # randomly choose the starting indices for the examples in the training batch
  idx = np.random.choice(n-seq_length, batch_size)

  #a list of input sequences for the training batch
  input_batch = [vectorized_songs[i : i+seq_length] for i in idx]
  #a list of output sequences for the training batch
  output_batch = [vectorized_songs[i+1 : i+seq_length+1] for i in idx]

  # x_batch, y_batch provide the true inputs and targets for network training
  x_batch = np.reshape(input_batch, [batch_size, seq_length])
  y_batch = np.reshape(output_batch, [batch_size, seq_length])
  return x_batch, y_batch
```

For each of these vectors, each index is processed at a single time step. So, for the input at time step 0, the model receives the index for the first character in the sequence, and tries to predict the index of the next character. At the next timestep, it does the same thing, but the RNN considers the information from the previous step, i.e., its updated state, in addition to the current input.


```python
x_batch, y_batch = get_batch(vectorized_songs, seq_length=5, batch_size=1)

for i, (input_idx, target_idx) in enumerate(zip(np.squeeze(x_batch), np.squeeze(y_batch))):
    print("Step {:3d}".format(i))
    print("  input: {} ({:s})".format(input_idx, repr(id_to_char[input_idx])))
    print("  expected output: {} ({:s})".format(target_idx, repr(id_to_char[target_idx])))
```

#3. The Recurrent Neural Network (RNN) model


We'll train our RNN using batches of song snippets from our dataset, which we have already generated.
The model we'll create will be based on LSTM architecture, where we will use state vector to maintain information about the temporal relationships between consecutive characters. The final output of the LSTM is then fed into a fully connected Dense layer where we'll output a softmax over each character in the vocabulary, and then sample from this distribution to predict the next character.

Layer 1: Embedding layer to transform indices into dense vectors of a fixed embedding size

Layer 2: LSTM with `rnn_units` number of units.

Layer 3: Dense (fully-connected) layer that transforms the LSTM output into the vocabulary size.

##Defining the model

```python
#nueva forma para definir el modelo

def build_model(vocab_size, embedding_dim, rnn_units, batch_size):
    # Creamos un modelo secuencial
    model = tf.keras.Sequential()

    # Agregamos una capa Input explícita para definir el batch_size
    model.add(tf.keras.layers.InputLayer(batch_input_shape=[batch_size, None]))

    # Agregamos la capa Embedding
    model.add(tf.keras.layers.Embedding(
        input_dim=vocab_size,
        output_dim=embedding_dim,
        name='embedding'
    ))

    # Agregamos la capa LSTM stateful
    model.add(tf.keras.layers.LSTM(
        units=rnn_units,
        return_sequences=True,
        recurrent_initializer='glorot_uniform',
        recurrent_activation='sigmoid',
        stateful=True,
        name='lstm'
    ))

    # Agregamos la capa Dense
    model.add(tf.keras.layers.Dense(
        units=vocab_size,
        name='dense'
    ))

    return model

# Creamos el modelo
model = build_model(len(vocab), embedding_dim=1024, rnn_units=2048, batch_size=32)
```

#4. Training the model: loss and training operations
To train our model, we can use a form of the crossentropy loss (negative log likelihood loss)i.e., sparse_categorical_crossentropy loss, as it utilizes integer targets for categorical classification tasks. We will want to compute the loss using the true targets -- the labels -- and the predicted targets -- the logits.

Computing loss from untrained model:

```python
### Defining the loss function ###

def compute_loss(labels, logits):
  loss = tf.keras.losses.sparse_categorical_crossentropy(labels, logits, from_logits=True)
  return loss


example_batch_loss = compute_loss(y, pred)

print("Prediction shape: ", pred.shape, " # (batch_size, sequence_length, vocab_size)")
print("scalar_loss:      ", example_batch_loss.numpy().mean())
```

Defining hyperparameters for training the model:

```python
### Hyperparameter setting and optimization ###

# Optimization parameters:
num_training_iterations = 50
batch_size = 10
seq_length = 100
learning_rate = 1e-3

# Model parameters:
vocab_size = len(vocab)
embedding_dim = 256
rnn_units = 1024

# Checkpoint location:
checkpoint_dir = './training_checkpoints'
checkpoint_prefix = os.path.join(checkpoint_dir, "my_ckpt")
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import time
import os
from tqdm.notebook import tqdm

### Hiperparámetros ###
embedding_dim = 1024 #antes 512
rnn_units = 2048 # antes 1024
batch_size = 32
seq_length = 100
learning_rate = 0.0003
num_training_iterations = 2000
reset_states_interval = 100

### Suponiendo que ya tienes: ###
# - 'vocab' definido
# - 'vectorized_songs' generado
# - 'get_batch' y 'compute_loss' definidos
vocab_size = len(vocab)

### Función para construir el modelo ###
def build_model(vocab_size, embedding_dim, rnn_units, batch_size):
    model = tf.keras.Sequential([
        tf.keras.layers.InputLayer(batch_input_shape=[batch_size, None]),
        tf.keras.layers.Embedding(input_dim=vocab_size, output_dim=embedding_dim, name='embedding'),
        tf.keras.layers.LSTM(units=rnn_units,
                             return_sequences=True,
                             recurrent_initializer='glorot_uniform',
                             recurrent_activation='sigmoid',
                             stateful=True,
                             name='lstm'),
        tf.keras.layers.Dense(units=vocab_size, name='dense')
    ])
    return model

### Checkpoints ###
checkpoint_dir = './training_checkpoints'
checkpoint_prefix = os.path.join(checkpoint_dir, 'my_ckpt.weights.h5')
if not os.path.exists(checkpoint_dir):
    os.makedirs(checkpoint_dir)

### Crear modelo y optimizador ###
model = build_model(vocab_size, embedding_dim, rnn_units, batch_size)
optimizer = tf.keras.optimizers.Adam(learning_rate)

### Función de entrenamiento ###
@tf.function
def train_step(x, y):
    with tf.GradientTape() as tape:
        y_hat = model(x)
        loss = compute_loss(y, y_hat)
    grads = tape.gradient(loss, model.trainable_variables)
    grads, _ = tf.clip_by_global_norm(grads, 5.0)  # Evita explosiones de gradiente
    optimizer.apply_gradients(zip(grads, model.trainable_variables))
    return loss

### Entrenamiento ###
history = []

if hasattr(tqdm, '_instances'):
    tqdm._instances.clear()

print("Comenzando el entrenamiento...")
for iter in tqdm(range(num_training_iterations)):
    x_batch, y_batch = get_batch(vectorized_songs, seq_length, batch_size)
    loss = train_step(x_batch, y_batch)
    loss_value = loss.numpy().mean()
    history.append(loss_value)

    if iter % 50 == 0:
        print(f"Iteración {iter}: Pérdida = {loss_value:.4f}")

    if iter % 100 == 0:
        model.save_weights(checkpoint_prefix)

    if iter % reset_states_interval == 0:
        model.get_layer('lstm').reset_states()


# Guardar pesos finales y modelo completo
model.save_weights(checkpoint_prefix)
print(f"Entrenamiento terminado. Pérdida final: {history[-1]:.4f}")

plt.figure(figsize=(10, 6))
plt.plot(history)
plt.title('Pérdida durante el entrenamiento')
plt.xlabel('Iteraciones')
plt.ylabel('Pérdida')
plt.grid(True)
plt.show()

model.save(os.path.join(checkpoint_dir, 'full_model32bs.keras'))
```

##El procedimiento de predicción

Paso 1: Inicializar una cadena de inicio "semilla" y el estado de la RNN, y establecer la cantidad de caracteres que queremos generar.

Paso 2: Usar la cadena de inicio y el estado de la RNN para obtener la distribución de probabilidad del siguiente carácter predicho.

Paso 3: Muestrear a partir de una distribución multinomial para calcular el índice del carácter predicho. Este carácter predicho se usa luego como la siguiente entrada para el modelo.

Paso 4: En cada paso temporal, el estado actualizado de la RNN se vuelve a alimentar al modelo, de modo que ahora tiene más contexto para hacer la siguiente predicción. Después de predecir el siguiente carácter, los estados actualizados de la RNN se vuelven a alimentar al modelo, lo cual es la forma en que aprende las dependencias de secuencia en los datos, ya que obtiene más información de las predicciones anteriores.

```python
### Prediction of a generated song ###
def generate_text_alternative(model, start_string, generation_length=1000):
    # Esta versión evita usar reset_states()

    # Convertir la cadena inicial a números (vectorizar)
    input_eval = [char_to_id[s] for s in start_string]
    input_eval = tf.expand_dims(input_eval, 0)

    # Cadena vacía para almacenar nuestros resultados
    text_generated = []

    # Crear un nuevo modelo con batch_size=1 para la generación
    # Esto evita la necesidad de reset_states()
    generation_model = build_model(len(vocab), embedding_dim=1024, rnn_units=2048, batch_size=1) #funciona bien
    #generation_model = build_model(len(vocab), embedding_dim=512, rnn_units=1024, batch_size=1) #funciona descente
    #generation_model = build_model(len(vocab), embedding_dim=2048, rnn_units=4096, batch_size=1)

    generation_model.set_weights(model.get_weights())

    from tqdm import tqdm
    for i in tqdm(range(generation_length)):
        # Evaluar las entradas y generar las predicciones del siguiente caracter
        predictions = generation_model(input_eval)

        # Eliminar la dimensión de batch
        predictions = tf.squeeze(predictions, 0)

        # Usar una distribución multinomial para la muestra
        predicted_id = tf.random.categorical(predictions, num_samples=1)[-1,0].numpy()

        # Pasar la predicción junto con el estado oculto anterior como la siguiente entrada al modelo
        input_eval = tf.expand_dims([predicted_id], 0)

        # Agregar el caracter predicho al texto generado
        text_generated.append(id_to_char[predicted_id])

    return (start_string + ''.join(text_generated))


generated_text = generate_text_alternative(model, start_string="(F3 _G)", generation_length=1000)
print(generated_text)
```
